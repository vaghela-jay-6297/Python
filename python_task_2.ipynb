Create String varible and assign some values then apply some string functions.

```
a = "asdfghjkl saasds asadedw"
print(a)  
```

```
print("capitalize -> ", a.capitalize())	#Capital only First letter of line
print("Upper -> ", a.upper()) 	#Uppercase All String
print("Lower -> ", a.lower()) 	#Lowercase All String
```

```
b = "HETAL hi hello"
print(b)
print("CaseFold -> ", b.casefold())  #Note : Upper case convert to Lower case But Lower Case is not convert to upper case

print("Center -> " , b.center(30)) 

print("Count(h) ->" , b.count("h"))	#count All h in above String

print("EndWith Fun -> " , b.endswith("o"))	# return in bool
print("StartWith fun ->" ,b.startswith("E"))
```

```
c = "abcd\tdef\tghig"
print(c)

print("Expand Tab -> " ,c.expandtabs(12))	#whitespace count with value like abcd(4) than whitespace(8)

print ("Find Character index(Position)->",c.find("d"))	#return index of character start with 0 like d=3

print("Find Last Character ->",c.rfind("g")) #find From Right side and last character
```

```
d = "12"
print(d)

print("alphanumerical -> ", d.isalnum()) #alphanumerical value(ex:'ab12') is true otherwise false for special symbols
print("Character ->", d.isalpha()) #check only character is return true otherwise false
```

```
f = "0.1234"  # 12 is decimal and digit
print(f)
print("Decimal -> ", f.isdecimal()) #check decimal value like 5,10,23,12 ....

print("Digit -> ", f.isdigit())
```

```
g = "    ABCDEF"
print(g)

print("Check Lower -> ",g.islower()) #check String is lower or not
print("Check Upper -> ",g.isupper()) #check String is upper or not

print("Convert to lower case -> ", g.lower()) 	#convert to lower case

print("Space -> ",g.isspace()) 	#null String like "  " is true

print("Join -> ",'a-b'.join("@123@"))	#join character first @ than a-b 1 than a-b
```

```
h = "     I   am   Jay jjjjjj  Vaghela       123    "
print(h)

print("Remove lest side Space ->" ,h.lstrip()) #remove lest side White Space
print("Remove all Space -> ", h.strip())

print("Replace Character -> ",h.replace("Jay","Iron Men"))

print("Split -> " ,h.split())	#divide particular words like I,am,Jay....
```

```
i = "HeLLo WoRld ! \n aaaaAAAAAaaaa \n bbbbbbbbbbbbbBBBBBBBBBBbbbbbbb"
#i = '''Dear <Name>, \n Yor Are selected for python job. \n you can join our company from 12 jan 2019 \n Thank you! ''' # mail template
print(i)
print("Split Lines -> ",i.splitlines())

print("Swap Case -> ",i.swapcase())

print("Index From Left To right -> " ,i.rindex("d")) #postiotion of 'd'


z = i.zfill(80)
print("Fill Zeros -> ",z)
print("Original Length of i:{}, Zfill Length:{}, Calculate diff(original String-after fill zeros):{} zeros filled ".format(len(i),len(z),len(z)-len(i)) )
```
