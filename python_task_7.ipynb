## Task: 7 Swap value using different different logics.
 

```
# 1st Logic(python provide easy way.)
def first_swap_value(a,b):
  print("Original Values are -> a = {}, b = {}".format(a,b))
  a,b = b,a
  return a,b

swap_a ,swap_b = first_swap_value(134,456)
print("Swaped Values are -> a = {}, b = {}".format(swap_a ,swap_b) )

```

```
# 2nd Logic(Using Temp Var)
def second_swap_value(a,b):
  print("Original Values are -> a = {}, b = {}".format(a,b))
  temp = a
  a = b
  b = temp
  return a,b

swap_a ,swap_b = second_swap_value(134,456)
print("Swaped Values are -> a = {}, b = {}".format(swap_a ,swap_b) )
```

```
# 3rd Logic(Using Temp Var)
def third_swap_value(a,b):
  print("Original Values are -> a = {}, b = {}".format(a,b))
  a = a - b
  b = a + b
  a = b - a
  return a,b

swap_a ,swap_b = third_swap_value(134,456)
print("Swaped Values are -> a = {}, b = {}".format(swap_a ,swap_b) )
```
