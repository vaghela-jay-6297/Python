## Task :6 Python Operators

Create Some Variables and assign some values than perform some python operator operations.

```
a = int(input("Enter Value of a -> "))
b = int(input("Enter Value of b -> "))

x = True
y = False

list1 = [1,2,3,4,5,7,8,9]
list2 = [list1,7,8,9,10]
list3 = [1,7,3,8,9]

print("x-> ",x)
print("y-> ",y)

print("List1-> ",list1)
print("List2-> ",list2)
print("List3-> ",list3)
```

### Arithemetic Operators

```
print("\t ################################  A. Arithmetic Operators  ################################")
print("\n Operators are ->\t addition(+),\t subtraction(-),\t multiplication(*),\t floting Division(/),\t Modulo(%),\t Power(**),\t integer division(//)",)
print("\n a is -> ",a,"\t b is -> ",b)

print("\n A + B -> ",a+b)
print("\n A - B -> ",a-b)
print("\n A * B -> ",a*b)
print("\n A / B -> ",a/b)
print("\n A ** B -> ",a**b)	#power 2**3 = 2*2*2
print("\n A // B -> ",a//b) #integer division 
```

### Relational(Comparison) Operators

```
print("\t ################################  B. Comparison(Relational) operators  ################################")
print("\n Operators are ->\t Compare(==),\t Greater than(>),\t Less than(<),\t Greater than or equal(>=),\t Less than or equal(<=),\t Not equal(!=)",)
print("\n a is -> ",a,"\t b is -> ",b)

print("\n A == B -> ",a==b)
print("\n A > B -> ",a>b)
print("\n A < B -> ",a<b)
print("\n A >= B -> ",a>=b)
print("\n A <= B -> ",a<=b)
print("\n A != B -> ",a!=b)	
```

### Logical Operators


```
print("\t ################################  C. Logical operators  ################################")
print("\n Operators are ->\t AND(and),\t OR(or),\t NOT(not)")

print("\n X is -> ",x,"\t Y is -> ",y)
print('\n x and y is',x and y)
print('\n x or y is',x or y)
print('\n not x is',not x)
```

### Assignment operators

```
print("\t ################################  D. Assignment operators  ################################")
print("\n Operators are ->\t =,\t +=,\t -=,\t *=,\t /=,\t %=,\t //=,\t **=,\t &=,\t |=,\t ^=,,\t >>=,\t <<=")

#like Example
print("\n x = 5,\t x += 5,\t x -= 5,\t x *= 5,\t x /= 5,\t x %= 5,\t x //= 5,\t x &= 5,\t x |= 5,\t x ^= 5,\t x >>= 5,\t x <<= 5")
```

### BitWise Operators (Work on Binary Digits)

```
print("\t ################################ E. BitWise Operators  ################################")
print("\n Operators are ->\n ~(Complement),\n &(And),\n |(or),\n <<(Left Shift),\n >>(Right Shift),\n ^(Xor)")
print("\n a is -> ",a,"\t b is -> ",b)

#bitwise operators are work on binary number or bits.
#always decimal numbers are convert to the binary number than apply operations.

print("----------  1. Complement(~) ---------- ")
print("\t Positive Complement Of A -> ",~a)	#for Positive number first find the binary number second add +1(2's) of a.
print("\t Negative Complement of A -> ",~-a)	#for negative number first find the binary number second find 1's com than add+1(2's) third again find 1's com on there answer.

print("\n ---------- 2. AND(&) ---------- ")
print("\t binary  -> ",bin(a),'\t',bin(b),'\t',bin(a&b))	#binary of A and B (true & true = true ,1 & 1 = True)
print("\t Decimal -> ",a,'\t',b,'\t \t',(a&b))	#print answer in decimal format

print("\n ---------- 3. OR(|) ---------- ")
print("\t binary  -> ",bin(a),'\t',bin(b),'\t',bin(a|b)) #(false | false = false , 0 | 0 = False)
print("\t Decimal -> ",a,'\t',b,'\t \t ',a|b)	#print answer in decimal format

print("\n ---------- 4. Left Shift(<<) ---------- ")
print("\t binary Left Shift -> ",bin(a),'\t',bin(a<<2)) #2 bits shift left side like 00001100 << 00110000
print("\t Decimal Left Shift -> ",a,'\t',a<<2) #print in decimal digit

print("\n ---------- 5. Right Shift(>>) ---------- ")
print("\t binary Right Shift -> ",bin(a),'\t',bin(a>>2)) #2 bits shift Right side like 00001100 >> 00000011
print("\t Decimal Right Shift -> ",a,'\t',a>>2) #print in decimal digit

print("\n ---------- 6. XOR(^) ---------- ")
print("\t binary Xor of A ^ B -> ",bin(a),'\t',bin(b),'\t',bin(a^b)) #in xor (1 ^ 0 = 1 , 0 ^ 1 = 1)
print("\t Decimal Xor of A ^ B -> ",a,'\t',b,'\t \t ',a|b) #print in decimal digit
```

### Special Operators

```
print("\t ################################ F. Special Operators  ################################")

print("\n ------------------ 1. Identity Operators ------------------")
#is and is not are the identity operators in Python. They are used to check if two values (or variables) are located on the same part of the memory. Two variables that are equal does not imply that they are identical.
print("\n \t Operators are ->\t is,\t is not")
print("\t\t Lists are -> ",list1,list2,list3)
print("\t\t a is -> ",a,"\t b is -> ",b)

#list1 and list2 are list. They are equal but not identical. It is because interpreter locates them separately in memory although they are equal.
print("\t\t list1 is list2 -> ",list1 is list2)
print("\t\t a is b -> ", a is b)

print("\n ------------------ 2. Membership operators ------------------")
#in and not in are the membership operators in Python. They are used to test whether a value or variable is found in a sequence (string, list, tuple, set and dictionary).
print("\n \t Operators are ->\t in,\t not in")
print("\t\t Lists are -> ",list1,list2,list3)

print("\t\t list1 in list2 -> ",list1 in list2)
print("\t\t list2 not in list3 -> ",list2 not in list3)
print("\t\t 5 in list1 -> ",5 in list1)
```
