### Task-1 Create Variable with Values And show Variable Memory Address.

```
print("***** Task -1 : Create Variable with Values And show Variable Memory Address. *****")
```

Create Variable and Assign Some values.


```
varA = 101 	#integer
varB = 'Jay' 	#String
varC = "a b c d e f g h i j k l m n o p q r s t u v w x y z" 	#String
varD = 10<20	#bool
varE = 10.345	#float
varF = None  #None
```

print value of variable and there maemory address with datatype.


```
print("Print All Variables Values With There Memory Address")
print("VarA ==== Value:{}, Address:{}, DataType:{} ".format(varA,id(varA),type(varA)) )
print("VarB ==== Value:{}, Address:{}, DataType:{} ".format(varB,id(varB),type(varB)) )
print("VarC ==== Value:{}, Address:{}, DataType:{} ".format(varC,id(varC),type(varC)) )
print("VarD ==== Value:{}, Address:{}, DataType:{} ".format(varD,id(varD),type(varD)) )
print("VarE ==== Value:{}, Address:{}, DataType:{} ".format(varE,id(varE),type(varE)) )
print("VarF ==== Value:{}, Address:{}, DataType:{} ".format(varF,id(varF),type(varF)) )
```
