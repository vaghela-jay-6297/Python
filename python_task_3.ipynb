## Task:3 Set(Collection Type) Operation

Create Set variable and assign Some values.

```
fruits = {"apple", "banana" , "cherry","pineapple","mango","peach","apple"} #Create Set
fruit={"lychees","blackberries","banana","mango","papaya","apple","cherry"}
fruit1={"apple","mango","cherry"}

print("fruits -> ",fruits ,'\n',"fruit -> ",fruit ,'\n',"fruit1-> ",fruit1)
```

```
# symmetric difference =  (a-b)+(b-a)
f = fruits.symmetric_difference(fruit)	#print the symmetric difference b/w fruits and fruit and add all difference values of both set
print("Difference -> ",f)
print("Difference -> ",fruits ^ fruit)
```

```
print("Subset(true/false) - > ", fruit1.issubset(fruit))	#check fruit1 is subset of fruit

print("SuperSet(true/false) -> ",fruits.issuperset(fruit))	#check fruits is superset of fruit

print("DisJoint Set  ->" , fruit.isdisjoint(fruits))		#compare all elements of set.all elements are diiferent than return true
```

```
fruits.add("12")		# Add values in set
fruits.add("13")
fruits.add("14")
print("fruits",fruits) 

fruit2.discard("apple")		#Remove Element From Set
print("Discard -> ",fruit2)

fruits.pop()
print("Pop Set Elements -> ",fruits) 	#remove set element
```

```
fruit2 = fruit1.copy()		#copy Of set
print("Copy -> ",fruit2)
```

```
print("Difference -> ", fruits.difference(fruit))	#a-b
print("Difference -> ", fruits - fruit)

print("\n")

print("Intersection -> ", fruits.intersection(fruit))	#same element in two sets is print
print("Intersection -> ", fruits & fruit) #using operator

print("\n")

print("Union -> ",fruits.union(fruit))	#union of two set combine all element of two set
print("Union -> ",fruits|fruit)	#using operator
```
