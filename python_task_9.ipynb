## Task:9 Create Some Patterns.

```
# Pattern One
row = int(input("Enter How Many Rows -> "))
for i  in range(0,row):
  for j in range(row):
    print("*",end=" ")
  print()
```

```
# Pattern Two
row = int(input("Enter How Many Rows -> "))
for i  in range(0,row):
  for j in range(i+1):
    print("*",end=" ")
  print()
```

```
# Pattern Three
row = int(input("Enter How Many Rows -> "))
for i  in range(0,row):
  for j in range(i+1):
    print(j+1,end=" ")
  print()
```

```
# Pattern four
n=5
for i in range(n):
  n = n-1
  for j in range(n,0,-1):
    print("-",end=" ")
  for k in range(i+1):
    print("#",end=" ")
  print()
```
