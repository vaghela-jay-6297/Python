## Task:8 Python if,else statements

```
# 1st program
print("***** Check Number is Odd or Even *****")
num = int(input("Enter Number -> "))
r = num % 2

if r == 0:  # if you can enter even number than excute this block
  print("Number is Even")
if r == 1:  # if you can enter odd number than excute this block
  print("Number is Odd")


if num % 2 == 0:  # if you can enter even number than excute this block
  print("Number is Even")
else:   # if you can enter odd number than excute this block
  print("Number is Odd")
```

```
# 2nd program
print("***** Find the Largest Number *****")
a = int(input("Enter 1st Number -> ")) 
b = int(input("Enter 2nd Number -> "))
c = int(input("Enter 3rd Number -> "))

large_num=0
if (a > b) and (a > c) :
  large_num = a
elif (b > a) and ( b > c ):
  large_num = b
else:
  large_num = c

print("Largest Number is -> ",large_num)
```

```
# 3rd program
print("***** Find Number is Positive or Negative *****")
no = int(input("Enter Number -> "))

if no == 0:
  print("Number is Zero")
elif no < 0:
  print("Number is Negative")
else:
  print("Number is Positive")
```
