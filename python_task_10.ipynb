## Task:10 Types of Function Arguments.

```
print("1)Formal Argument \n2)Actual Argument")
print("-> Actual Arguments have four types: \n i)Position Arguments \t ii)Default Arguments \n iii)Keyword Arguments \t iv)Variable Length Arguments")
```

```
#Position Argument
def position_argument(name,age):
  print("My Name is '{}' ".format(name))
  print("My Age is '{}' ".format(age))

position_argument("Jay Vaghela",23)
```

```
#Default Argument
def position_argument(name,age=0.0):  #if use not pass the value of age than automatically Get default value of age is 0.0.
  print("My Name is '{}' ".format(name))
  print("My Age is '{}' ".format(age))

position_argument("Default Argument") 
print()
position_argument("Jay Vaghela",23)
```

```
#Keyword Argument
def position_argument(name,age):
  print("My Name is '{}' ".format(name))
  print("My Age is '{}' ".format(age))

position_argument(age=23,name="Jay Vaghela") # pass value with keywords
```

```
#Variable Length Argument
def Total_sum(a,*b): #here a is fix value but b have multiple values.
  sum = a
  for i,item in enumerate(b):
    sum = sum + item
  print("Toatl Sum is '{}' ".format(sum))

Total_sum(10,20,23,34,78,45,80) # here a=10 is fixed but b have multiple values such as 20,23,34,...,
Total_sum('A','B','C','D','E','F')
```

```
#Keywords Variable Length Argument(kwargs)
def Total_sum(name,**data): #here a is fix value but b have multiple values.
  print("Employee name '{}' ".format(name))
  print("\nSome Others Details Of {} is listed Below:".format(name))
  for i,item in data.items():
    print("{} is {}.".format(i,item))

Total_sum(name='Vaghela Jay',age=23,Post='Developer',gender='M',dob='06-02-1997')
```
