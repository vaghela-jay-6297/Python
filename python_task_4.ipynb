## Task:4 Create Some List,Tuple & apply Some Functions On list,tuple.

```
tup1=(10,20,30)
tup2=(15,"ldjn",22.6)

print("Tup1 ->",tup1)
print("Tup2 ->",tup2)
```

```
print("Count-> ",tup1.count(10))
print("Index-> ",tup2.index(22.6))
```

```
car=['maruti','bmw','audi','mercedes','walkswagen','porche','bmw']
list1=['A','B','C','D','E']
list2 = list1
set1={0.45,1.89,6.78,8,90,1,243}
fruit=['apple','banana','mango','pineapple',tup1,tup2,list1,set1]

print(" ##########  Print Lists nd Tuples  ##########")
print("List is -> ",car)	#list
print("List is -> ",fruit)	#list
print("Tuple is -> ",tup1 ,'\n' "Tuple is -> ",tup2)
print("Set is -> ",set1)
```

```
print("list2 -> ",list2)
for i in list2:
	print(i)
```

```
print("\n print items by there index -> ", car[0],'\t',car[1:3],'\t',car[3:]) #print element by index
print("print items from right to left -> ",car[-3],'\t',car[:-1]) #print element right to left (last to first)
print("Nested list element -> ",fruit[4][1],'\t',fruit[5][2],'\t',fruit[6][4])
print("repeat -> ",list1*3 )
aList = list(tup1) #tuple convert to list
print("Tuple Convert To list -> ",aList)
```

```
bList = list(set1)	#set convert to list
print("Set Convert to list -> ",bList)
```

```
list1.append("F")	#append the element at last position of list
print("Append List -> ",list1)

list1.insert(2, "ABCD")	#insert element with specific index
print("insert element with index List -> ",list1)

list1.extend(['G','H','I'])
print("Extend element in list -> ",list1)
```

```
list1.remove("ABCD")	#delete element of list but only one element at a time
print("Delete element of List -> ",list1)

list1.pop(3)	#delete element by index. Stack Concept
print("After Poped Value -> ",list1)

list1.pop()	#remove Last element
print("Poped Value -> ",list1)
```

```
list1.sort()	#not suppoted for tuple and set
print("Sorted List -> ",list1)

list1.reverse()	#reverse list
print("Reverse List -> ",list1)
```

```
list2 = [10,20,70,34,67,35,10,20]
print("Minimum value of list -> ",min(list2))	#build in fun
print("Maximum value of list -> ",max(list2))
print("Sum of list -> ",sum(list2))
```

```
pow2 = [2 ** x for x in range(10)]
print("Power of 2 -> ",pow2)

pow3 = [2 ** x for x in range(10) if x > 5]
print("Power -> ",pow3)
```

```
print("A in list -> ",'A' in list1)

print("using for Loop -> ")
for i in ['a','b','c']:
	print(i)
```

```
#del list1	# entire list & tuple Also
del list1[1:]	#pass index of list elements
print("Delete Multiple Elements in list -> ",list1)
```
